```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np 
```

```python
#Loading Esport earning data found on Kaggle (Esports Earnings 1998 - 2020)
csv_file = "../esportfilecsv/GeneralEsportData.csv"
esport = pd.read_csv(csv_file, encoding='ISO-8859-1')
esport.head()
```

```python
#Esport index reset 
esport = esport.reset_index()
esport.head()
```

```python
#Setting Index to its own column
esport = esport.rename(columns = {'index':'id'})
esport.set_index('id', inplace = True)
esport.head()
```

```python
#Importing Data from Data.word(Video Games Global Sales in Volume (1983-2017))
csv_file2 = "../vgsalesGlobale.csv"
game = pd.read_csv(csv_file2, encoding='ISO-8859-1')
game.head()
```

```python
#Keeping the needed colums from the game table 
game_data = game[['Name', 'Global_Sales']].copy()
game_data.head()
```

```python
#Rename the columns to match the colums from PGadmin tables 

esport = esport.rename(columns = {'Game':'game', 
                                  'ReleaseDate':'releasedate',
                                  'Genre':'genre', 
                                  'TotalEarnings':'totalearnings',
                                  'OnlineEarnings':'onlineearnings', 
                                  'TotalPlayers':'totalplayers',
                                  'TotalTournaments':'totaltournaments'})

esport
```

```python
#Rename the columns to match the colums from PGadmin tables 

game_data = game_data.rename(columns= {'Name':'name',
                             'Global_Sales':'global_sales' })
game_data
```

```python
#Converting numbers from the game_data dataframe
game_data['global_sales'] = game_data['global_sales']*(10**6)
game_data
```

```python
#groupby names and do the summation of the global sales across game plateforms
groupgame = game_data.groupby(['name'],as_index = False).sum()
groupgame.head()
```

```python
#Set index to game_data dataframe
groupgame = groupgame.reset_index()
groupgame.head()
```

```python
#Changing column name from index to id 
#Set id column to index 
groupgame = groupgame.rename(columns = {'index':'id'})
groupgame.set_index('id', inplace = True)
groupgame.head()
```

```python
#Connect to databases
rds_connection_string = "username:password@localhost:5432/Esport_rev_earn"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Create variable to connect 
conn = engine.connect()
```

```python
#Check tables from the database
engine.table_names()
```

```python
#Import jupyter notebook esport dataframe into esport pgadmin table
esport.to_sql(name='esport', con=engine, if_exists='append', index=True)
```

```python
#Import jupyter notebook game_data dataframe into pgadmin game table
groupgame.to_sql(name='game', con=engine, if_exists='append', index=True)
```

```python
#Check tables from the database
engine.table_names()
```

```python
#Read Merged_esport table from database 
merged_esport = pd.read_sql('select * from merged_esport', conn)
merged_esport.head()
```
